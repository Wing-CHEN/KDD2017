```python
#reform the date
def reform_date(date):
    _date_ = re.split('\/',date)
    if len(_date_[1])==1:
        _date_[1]='0'+_date_[1]
    if len(_date_[2])==1:
        _date_[2]='0'+_date_[2]
        
    return _date_[0]+'-'+_date_[1]+'-'+_date_[2]
```

```python
#Creat time window
def get_time_window(time):
    time = datetime.strptime(time,"%Y-%m-%d %H:%M:%S")
    time_window_minute = int(math.floor(time.minute/20)*20)
    time_window_start = datetime(time.year,time.month,time.day,time.hour,
                                 time_window_minute,0)
    time_window_end = time_window_start + timedelta(minutes=20)
    time_window ='[' + str(time_window_start) + ',' + str(time_window_end) +")"
    return time_window
```

```python
#Creat time hour
def get_time_hour(time):
    time = datetime.strptime(time,"%Y-%m-%d %H:%M:%S")
    time_hour = int(math.floor(time.hour/3)*3)
    return time_hour
```

```python
volume_raw_training_data['time_window'] = volume_raw_training_data['time'].apply(get_time_window)
volume_raw_training_data['time_hour'] = volume_raw_training_data['time'].apply(get_time_hour)
```

```python
def get_wd(wind_direction):
    wd_ =np.arange(0,371.25,22.5)
    if wind_direction != 999017:
        temp_list = [math.fabs(wind_direction-wd) for wd in wd_]
        wd_index = temp_list.index(min(temp_list))
    else:
        return 'no_direction' 
    
    switcher ={0:'N',1:'NNE',2:'NE',3:'ENE',4:'E',5:'ESE',6:'SE',7:'SSE',8:'S',9:'SSW',
               10:'SW',11:'WSW',12:'W',13:'WNW',14:'NW',15:'NNW',16:'N'}
    return switcher.get(wd_index,'no_direction')
   
    
```

```python
#redifine time window
def count_time(time_window):
    time_ = datetime.strptime(re.split('\[|\ |\,|\)',time_window)[1],'%H:%M:%S')
    return int((time_.hour*3 )+ (time_.minute/20))
    
    
volume_collected_training_data['time_window_'] = volume_collected_training_data['time_window'].apply(count_time)
```

```python
#redifine date
volume_collected_training_data['date'] = volume_collected_training_data['date'].apply(lambda x: int(re.split('\-',x)[2]))
```

```python
#redifine wind_direction
def redifine_wd(wind_direction):
    switcher ={
        'N':0,
        'NNE':1,
        'NE':2,
        'ENE':3,
        'E':4,
        'ESE':5,
        'SE':6,
        'SSE':7,
        'S':8,
        'SSW':9,
        'SW':10,
        'WSW':11,
        'W':12,
        'WNW':13,
        'NW':14,
        'NNW':15,
        'N':16,
        'no_direction':17,
    }
    return switcher.get(wind_direction)
volume_collected_training_data['wind_direction'] = volume_collected_training_data['wind_direction'].apply(redifine_wd)
```
