```python
data.info() 
```

```python
tollgate_1_entry = data[(data.tollgate_id == 1)&(data.direction == 0)]
tollgate_2_entry = data[(data.tollgate_id == 2)&(data.direction == 0)]
tollgate_3_entry = data[(data.tollgate_id == 3)&(data.direction == 0)]

tollgate_1_exit = data[(data.tollgate_id == 1)&(data.direction == 1)]
tollgate_3_exit = data[(data.tollgate_id == 3)&(data.direction == 1)]
```

```python
tollgate_3_exit.head()
```

```python
#extract date index
tollgate_1_entry_index = [tollgate_1_entry.iat[x,1].split(' ')[0][1:11] for x in range(len(tollgate_1_entry))]
tollgate_2_entry_index = [tollgate_2_entry.iat[x,1].split(' ')[0][1:11] for x in range(len(tollgate_2_entry))]
tollgate_3_entry_index = [tollgate_3_entry.iat[x,1].split(' ')[0][1:11] for x in range(len(tollgate_3_entry))]

tollgate_1_exit_index = [tollgate_1_exit.iat[x,1].split(' ')[0][1:11] for x in range(len(tollgate_1_exit))]
tollgate_3_exit_index = [tollgate_3_exit.iat[x,1].split(' ')[0][1:11] for x in range(len(tollgate_3_exit))]

```

```python
#get new form
new_tollgate_1_entry = tollgate_1_entry.set_index([tollgate_1_entry_index])
new_tollgate_2_entry = tollgate_2_entry.set_index([tollgate_2_entry_index])
new_tollgate_3_entry = tollgate_3_entry.set_index([tollgate_3_entry_index])

new_tollgate_1_exit = tollgate_1_exit.set_index([tollgate_1_exit_index])
new_tollgate_3_exit = tollgate_3_exit.set_index([tollgate_3_exit_index])


new_tollgate_1_entry
```

```python
from datetime import datetime
import time
import re
rng = pd.date_range('19/09/2016','17/10/2016',freq='D')
```

```python
#tollage No.1  entry
show_tollgate_1_entry = pd.DataFrame()
for date_10 in rng:
    new_tollgate_1_entry_in_oneday = new_tollgate_1_entry[new_tollgate_1_entry.index == date_10.strftime('%Y-%m-%d')]
    time_index_10= [re.split('\[|\,|\)|\ ',i)[2] for i in new_tollgate_1_entry_in_oneday.time_window.values]
    temp_dateframe_10 = pd.DataFrame(new_tollgate_1_entry_in_oneday.volume.values,index=time_index_10,columns=[date_10.strftime('%Y-%m-%d')])
    show_tollgate_1_entry = pd.concat([show_tollgate_1_entry,temp_dateframe_10],axis=1)
#show_tollgate_1_entry
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y2)
# plot the second curve in this figure with certain parameters
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
# set x limits
plt.xlim((-1, 2))
plt.ylim((-2, 3))

# set new ticks
new_ticks = np.linspace(-1, 2, 5)
plt.xticks(new_ticks)
# set tick labels
plt.yticks([-2, -1.8, -1, 1.22, 3],
           ['$really\ bad$', '$bad$', '$normal$', '$good$', '$really\ good$'])
# to use '$ $' for math text and nice looking, e.g. '$\pi$'

# gca = 'get current axis'
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
# ACCEPTS: [ 'top' | 'bottom' | 'both' | 'default' | 'none' ]

ax.spines['bottom'].set_position(('data', -1))
# the 1st is in 'outward' | 'axes' | 'data'
# axes: percentage of y axis
# data: depend on y data

ax.yaxis.set_ticks_position('left')
# ACCEPTS: [ 'left' | 'right' | 'both' | 'default' | 'none' ]

ax.spines['left'].set_position(('data',0))
plt.show()
```
